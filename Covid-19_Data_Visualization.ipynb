```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Latest Covid-19 Data As of August 25

```python
data = pd.read_csv('Latest_Covid-19_India_Status.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.heatmap(data.corr())
```

Total Cases Reported in each State

```python
plt.figure(figsize = (10, 10))
plt.title('Total cases for each state in India')
sns.barplot(x=data["Total Cases"], y=data["State/UTs"])
```

Maharashtra has reported the most number of covid cases.
Andaman And Nicobar has the least among the lot

'Total Active cases for each state in India

```python
plt.figure(figsize = (10, 8))
plt.title('Total Active cases for each state in India')
sns.barplot(x=data["Active"], y=data["State/UTs"])
```

Kerela Presently has the most number of active covid Cases

```python
plt.figure(figsize = (10, 8))
plt.title('Total cases for each state in India')
sns.barplot(x=data["Deaths"], y=data["State/UTs"])
```

Maharashtra reported the most number of Deaths.
Daman And Diu Reported the least, only 4 deaths.

Total Cases Vs Deaths 

```python
plt.figure(figsize = (10, 8))
sns.lineplot(x=data["Deaths"], y=data["Total Cases"])
```

Discharged vs Total Cases Trend

```python
plt.figure(figsize = (10, 8))
sns.lineplot(x=data["Discharged"], y=data["Total Cases"])
```

We can clearly see a linear trend between Total Cases and the Total Number of Discharged patients

Deaths vs Discharged trend

```python
plt.figure(figsize = (12, 14))
sns.lineplot(x=data["Deaths"], y=data["Discharged"])
```

```python
plt.figure(figsize = (10, 8))
plt.title('Total Active cases for each state in India')
sns.barplot(x=data["Death Ratio (%)"], y=data["State/UTs"])
```

Punjab has the highest Death Ratio(%) among all states.
Dadra And Nagar Haveli and Daman & Diu have the least Death ratio.
Death Ratio refers to the ratio of total number of deaths to the total cases
