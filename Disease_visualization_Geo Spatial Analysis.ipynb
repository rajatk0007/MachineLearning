```python
import numpy as np
import pandas as pd
import folium
```

```python
df1 = pd.read_csv('Latest_Covid-19_India_Status.csv')
```

```python
df1.head()
```

```python
df2 = pd.read_csv('india_lat_and_long.csv')
```

```python
df2.head()
```

```python
df = df1.merge(df2,how="outer")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
centre_lat = df.mean().Latitude
centre_long = df.mean().Longitude
```

```python
map1 = folium.Map(location=[centre_lat, centre_long], zoom_start=4.4,tiles="Stamen Toner")
```

```python
for state in df["State/UTs"].unique():
    state = df[df["State/UTs"]==state]
    folium.CircleMarker(
        location=[state.Latitude, state.Longitude],
        radius = float(state["Total Cases"].values[0]/100000),
        popup="Total Cases : %s"%state["Total Cases"].values[0],
        tooltip = state["State/UTs"].values[0],
        color="blue",
        fill_color="blue"
    ).add_to(map1)
map1
```

In Total Cases Tally Maharashtra exceeds each State

```python
map2 = folium.Map(location=[centre_lat, centre_long], zoom_start=4.4,tiles="Stamen Toner")
```

```python
for state in df["State/UTs"].unique():
    state = df[df["State/UTs"]==state]
    folium.CircleMarker(
        location=[state.Latitude, state.Longitude],
        radius = float(state["Deaths"].values[0]/10000),
        popup="Total Deaths : %s"%state["Deaths"].values[0],
        tooltip = state["State/UTs"].values[0],
        color="red",
        fill_color="red"
    ).add_to(map2)
```

```python
map2
```

Maharashtra tops the charts in Total Deaths reported as well
